```python
import sys
sys.path += ['..']

import smbh

import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1.35e4
B = -1.48
H = 7540
K = 2.4e5
e = np.deg2rad(215)
```

```python
def Vm(n):
    global A, B
    return A * n ** 2 * (1 - 4 * n) ** 0.5 * (1 + B * n)

def Vpa(q, n, a1, a2, theta1, theta2):
    global H
    return H * (n ** 2 / (1 + q)) * (a1 * np.cos(theta1) - q * a2 * np.cos(theta2))

def Vpe(q, n, a1, a2, theta1, theta2, phi1, phi2):
    global K
    v = a1 * np.sin(theta1) * np.cos(phi1) - q * a2 * np.sin(theta2) * np.cos(phi2)
    return -K * n ** 3 / (1 + q) * v
    
def Vrecoil(q, a1, a2, theta1, theta2, phi1, phi2):
    global H, e
    n = q / (1 + q) ** 2
    
    vm = Vm(n)
    vpa = Vpa(q, n, a1, a2, theta1, theta2)
    vpe = Vpe(q, n, a1, a2, theta1, theta2, phi1, phi2)
    
    vr = vm ** 2 + vpa ** 2 + vpe ** 2 + 2 * vpe * (vm * np.cos(e) + vpa * np.sin(e))
    return vr ** 0.5

def getVMaxMin(qs, n = 500, theta = 0):
    A1 = np.random.rand(n) * 0.9
    A2 = np.random.rand(n) * 0.9

    theta1 = np.random.rand(n) * np.pi * theta
    theta2 = np.random.rand(n) * np.pi * theta

    phi1 = 2 * np.pi * np.random.rand(n)
    phi2 = 2 * np.pi * np.random.rand(n)

    vrs = np.zeros((n, len(qs)))
    vmax = 0
    pmax = -1
    vmin = 1e6
    pmin = -1

    for [i, (a1, a2, t1, t2, p1, p2)] in enumerate(zip(A1, A2, theta1, theta2, phi1, phi2)):
        vrs[i] = Vrecoil(qs, a1, a2, t1, t2, p1, p2)
        vmax_ = vrs[i].max()

        if vmax_ > vmax:
            pmax = i
            vmax = vmax_

        if vmax_ < vmin: 
            pmin = i
            vmin = vmax_
            
    return vrs[pmax], vrs[pmin]
```

```python
qs = np.logspace(-2, 0)
```

```python
v0 = getVMaxMin(qs, n = 500, theta = 0)
v1 = getVMaxMin(qs, n = 500, theta = 1)
```

```python
fig, axes = plt.subplots(1, 2, sharey = True, figsize = (8, 4.5))

for (ax, v) in zip(axes, (v1, v0)):
    ax.plot(qs, v[0])
    ax.plot(qs, v[1])
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid()
    
    ax.set_xlabel('$q = m_2 / m_1$')
    ax.set_xlim(qs.min(), qs.max())
    
axes[0].text(0.02, 1.5e3, r'$\theta_{1, 2}$ random')
axes[1].text(0.02, 1.5e3, r'$\theta_1 = \theta_2 = 0$')
axes[0].set_ylabel('Kick speed (km/s)')

axes[1].legend(['$v_{max}$', '$v_{min}$'])

fig.subplots_adjust(wspace = 0.1)

fig.savefig('Tanaka.png', dpi = 300)
plt.show()
```
