```python
import sys
sys.path += ["../../.."]

import smbh
import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ["return_1_%d.dat" % i for i in range(50)]
results = [smbh.Results(name) for name in names]

n = 5
skip = int(len(results) / n)
colors = smbh.getColors(n)

fig, axes = smbh.plotPhaseSpaces(results, max_points = 500)

lines = [axes[2].plot([], [], c = c) for c in colors]

labels = ["$fs = %.2f$" % result.STELLAR_FRACTION for result in results[:: skip]]

plt.subplots_adjust(wspace = 0.0)

smbh.coolLegend(axes[2], labels, loc = 'center left', bbox_to_anchor = (-0.25, 0.8), ncol = 1)

# fig.tight_layout()

fig.savefig('PhaseSpace_scape.png', dpi = 300)
plt.show()
```
